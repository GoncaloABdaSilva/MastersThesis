```python
from ultralytics import YOLO

model = YOLO('yolo11n-seg.pt')

model.info()

results = model.train(data='crack-seg.yaml', imgsz=(960, 720), epochs=30, batch=8)
```

```python
import os

model = YOLO('/kaggle/working/runs/segment/train/weights/best.pt')

folder = "/kaggle/input/resampled-test-cracktree260/test_cracktree260/img"
image_files = sorted([f for f in os.listdir(folder)])
image_paths = [os.path.join(folder, f) for f in image_files]

results = model.predict(image_paths)

for result in results:
    result.show()
    fname = os.path.basename(result.path)
    save_path = os.path.join('/kaggle/working', fname)
    result.save(save_path )
```
